```python
import pandas as pd
```

```python
maize=pd.read_csv('/Users/Mr.Jay/ai-project/maize_yield_dataset.csv')
```

```python
maize.head()
```

```python
maize_data=['Region','Rainfall (mm)','Temperature (°C)','Soil pH','Soil Nutrients (N-P-K)','Pest & Disease Prevalence (%)','Previous Year Yield (kg/ha)']
```

```python
maize=maize[maize_data]
```

```python
maize.head()
```

```python
import matplotlib.pyplot as plt
```

```python
#HBar Chart of rainfall against previous year yield
plt.bar(maize['Previous Year Yield (kg/ha)'],maize['Rainfall (mm)'])
# Setting the x and y labels
plt.ylabel('Rainfall (mm)')

plt.xlabel('Previous Year Yield (kg/ha)')

plt.show()
```

```python
# histogram of rainfall

plt.hist(maize['Rainfall (mm)'])

plt.title("Histogram")

# Adding the rainfall

plt.show()
```

```python
# Scatter plot with rainfall against yield

plt.plot(maize['Rainfall (mm)'])

plt.plot(maize['Previous Year Yield (kg/ha)'])

# Adding Title to the Plot

plt.title("Scatter Plot")

# Setting the X and Y labels

plt.xlabel('Rainfall (mm)')

plt.ylabel('Previous Year Yield (kg/ha)')

plt.show()
```

```python
# importing packages

import seaborn as sns

# reading the database

sns.histplot(x='Rainfall (mm)', data=maize, kde=True, hue='Previous Year Yield (kg/ha)')

plt.show()
```
